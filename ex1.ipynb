O dataset MNIST foi proposto em 1998 pra a identificação de n⁰ escritos à mão. Trata-se de um problema de classificação de imagens, uma vez q atribui uma label única à totalidade de uma imagem. É um dos + conhecidos datasets q representa um problema q é tipicamente fácil pra os humanos de resolver, mas difícil pra a máquina, sendo representado por uma BD de dígitos escritos à mão c um conj de treino composto por 60k ex (train.csv), e um conj de teste composto por 10k ex (test.csv). Cd dígito é representado por uma matriz 28x28, totalizando 784 valores de pixels em escala de cinzento (features), q variam entre 0 (branco) e 255 (preto). A 1⁰ coluna deste dataset (label) contém um valor entre 0 e 9 q identifica o dígito representado.

1. Utilizando H2O Flow, treine um modelo de DL (redes neuronais) pra o dataset acima apresentado.
<br>
    a) Teste diferentes configs de validação. Utilize cm validação o dataset de test, ou um crossfold c um valor de N-Fold à sua escolha

```python
#  pip install h2o
import h2o
from h2o.estimators import H2ODeepLearningEstimator
```

```python
# Iniciar o sv H2O
h2o.init()
```

```python
# Importar datasets
train = h2o.import_file("dataset/mnist_train.csv")
test = h2o.import_file("dataset/mnist_test.csv")
```

```python
# A 1⁰ coluna é a label
train[0] = train[0].asfactor()  # converte pra classificação
test[0] = test[0].asfactor()   # converte pra classificação

# Separar colunas
x = train.columns[1:]  # features (pixels)
y = train.columns[0]   # label
```

```python
# Criar o modelo de DL
# Opção A: validação c test.csv
model_a = H2ODeepLearningEstimator(
    activation="Rectifier",
    hidden=[128, 64],
    epochs=10,
    stopping_metric="misclassification",
    seed=1234
)

model_a.train(x=x, y=y, training_frame=train, validation_frame=test)
```

```python
# Opção B: validação c CrossFold
model_b = H2ODeepLearningEstimator(
    activation="Rectifier",
    hidden=[128, 64],
    epochs=10,
    stopping_metric="misclassification",
    nfolds=5,
    seed=1234
)

model_b.train(x=x, y=y, training_frame=train)
```

```python
# Avaliar o modelo
model_a.model_performance(test_data=test) # pra modelo c validação externa
model_b.cross_validation_metrics_summary() # pra modelo c cross-validation
```

    b) Analise as diferenças

```python
# Análise comparativa entre validação com test.csv e cross-validation

print("===== Modelo A: Validação com test.csv =====")
perf_a = model_a.model_performance(test_data=test)
print("Accuracy top-1 (mais comum):", perf_a.hit_ratio_table().as_data_frame().iloc[0, 1])
print("Erro (logloss):", perf_a.logloss())
print("Confusion Matrix:")
perf_a.confusion_matrix().show()

print("\n===== Modelo B: Cross-validation (5 folds) =====")
# Nota: model_performance() não se aplica a cross-validation diretamente
# Por isso usamos os summary metrics da validação cruzada
print(model_b.cross_validation_metrics_summary())

# Obter a tabela resumo de métricas da cross-validation
summary_df = model_b.cross_validation_metrics_summary().as_data_frame()

# Filtrar apenas a linha da acurácia
accuracy_line = summary_df[summary_df[''].str.strip() == 'accuracy']
print("Cross-validation Accuracy média (modelo B):", accuracy_line['mean'].values[0])


# Comentário sobre as diferenças:
print("\n===== Análise: =====")
print("""
O Modelo A teve uma ligeiramente melhor performance em termos de acurácia e logloss quando avaliado com o conjunto test.csv.

O Modelo B, ao usar validação cruzada (nfolds=5), forneceu uma métrica mais estável e confiável ao longo de diferentes subconjuntos do treino — o que é ótimo para avaliar generalização.

As diferenças são pequenas, indicando que o modelo é consistente e não está a sofrer overfitting significativo.

👉 Portanto, a validação cruzada é ótima durante o desenvolvimento e afinação, mas o teste com dados reais (test.csv) continua essencial para avaliar a generalização final.
""")
```
